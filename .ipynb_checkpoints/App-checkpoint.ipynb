```python
model = torch.hub.load('ultralytics/yolov5', 'custom', path='yolov5/runs/train/exp2/weights/best.pt', force_reload=True)
```

```python
# img = os.path.join('data', 'images', 'train', '92.png')
img = Image.open('data/images/cow(3).jpeg')
```

```python
# img = os.path.join('test', 'cow_and_person.jpg')
img = Image.open('test/cow_test.jfif')
```

```python
for filename in os.listdir('data/images/validate'):
    f = os.path.join('data/images/validate', filename)
    # checking if it is a file
    if os.path.isfile(f):
        img = Image.open(f)

        results = model(img)
        results.print()

        width, height = img.size

        right = (width * 25) // 100
        left = (width * 25) // 100
        top = (height * 25) // 100
        bottom = (height * 25) // 100
  
        new_width = width + right + left
        new_height = height + top + bottom
  
        i = Image.new(img.mode, (new_width, new_height), (255, 255, 255))
  
        i.paste(img, (left, top))

        results = model(i)
        results.print()

        %matplotlib inline 
        plt.imshow(np.squeeze(results.render()))
        plt.show()
```
